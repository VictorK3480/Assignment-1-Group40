```python
import json
from pathlib import Path
import gurobipy as gp
from gurobipy import GRB
import matplotlib.pyplot as plt

# Load data
base = Path("data/question_1b")

with open(base / "DER_production.json") as f:
    pv_ratio = json.load(f)[0]["hourly_profile_ratio"]

with open(base / "usage_preference.json") as f:
    usage = json.load(f)[0]
ref_load = usage["load_preferences"][0]["hourly_profile_ratio"]
lambda_discomfort = 1.5

with open(base / "bus_params.json") as f:
    bus = json.load(f)[0]
b = bus["energy_price_DKK_per_kWh"]
s = bus["energy_price_DKK_per_kWh"]
GE = bus["export_tariff_DKK/kWh"]
GI = bus["import_tariff_DKK/kWh"]

hours = range(24)
pv = {i: pv_ratio[i] for i in hours}

# Build model
m = gp.Model()
m.Params.LogToConsole = 0

x = m.addVars(hours, lb=0)
P_in  = m.addVars(hours, lb=0)
P_out = m.addVars(hours, lb=0)
ell = m.addVars(hours, lb=0)
u = m.addVars(hours, lb=0)

# Energy balance
for i in hours:
    m.addConstr(ell[i] == x[i] + P_in[i] - P_out[i])
    m.addConstr(x[i] <= pv[i])
    m.addConstr(u[i] >= ell[i] - ref_load[i])
    m.addConstr(u[i] >= ref_load[i] - ell[i])

# Objective
m.setObjective(sum(
    P_out[i]*(s[i] - GE) -
    P_in[i]*(b[i] + GI) -
    lambda_discomfort*u[i]
    for i in hours
), GRB.MAXIMIZE)

m.optimize()

# Extract dual _i
balance_constrs = m.getConstrs()[:24]   # first 24 constraints are balance
lambda_i = [c.Pi for c in balance_constrs]

# Compute MWTP & MOC
MWTP = [lam - GI for lam in lambda_i]
MOC  = [lam + GE for lam in lambda_i]

# Plot
plt.figure(figsize=(14,5))   # wider
plt.plot(hours, MWTP, marker='o', label="Demand curve (MWTP)")
plt.plot(hours, MOC,  marker='o', label="Supply curve (MOC)")
plt.xlabel("Hour")
plt.xticks(hours, [str(i+1) for i in hours])
plt.ylabel("Price (DKK/kWh)")
plt.title("Hourly Demand & Supply Curves from KKT dual variables")
plt.grid(True)
plt.legend()
plt.show()
```
